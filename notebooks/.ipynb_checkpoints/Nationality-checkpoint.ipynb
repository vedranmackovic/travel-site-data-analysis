```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Travel details dataset.csv")

df = df.dropna(subset=['Traveler name', 'Traveler nationality'])

df['unique_traveler'] = df['Traveler name'].str.strip() + "-" + df['Traveler nationality'].str.strip()

nationality_counts = df.drop_duplicates(subset='unique_traveler')['Traveler nationality'].value_counts()

top_n = 10
top_nationalities = nationality_counts.head(top_n)
others_total = nationality_counts.iloc[top_n:].sum()

labels = list(top_nationalities.index)
values = list(top_nationalities.values)

if others_total > 0:
    labels.append("Others")
    values.append(others_total)

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops=dict(width=0.4, edgecolor='white')  # donut style
)

ax.set_title("Traveler Nationalities (Top 10)", fontsize=14)
plt.tight_layout()
plt.show()
```
